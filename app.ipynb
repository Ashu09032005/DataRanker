```python
X = table_list.drop('Result', axis=1)  # All features
y = table_list['Result']              # Target
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=True, handle_unknown='ignore')
```

```python
# Fit and transform, but DO NOT convert to dense array
X_encoded_sparse = encoder.fit_transform(table_list[[
    'CycleName', 'TestPriority', 'Defect1', 'Defect2', 'Defect3', 'Defect4',
    'DefectStatus1', 'DefectStatus2', 'DefectStatus3', 'DefectStatus4',
]])
```

```python
print(X_encoded_sparse)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded_sparse, y_encoded,test_size=0.20,random_state=42)
```

```python
##Model training

from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(multi_class='ovr')
```

```python
logistic.fit(X_train, y_train)
y_pred=logistic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

HYPERPARAMETER TUNING

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,0.1,0.01]
solver=['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
params={'penalty':penalty,'C':c_values,'solver':solver}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model=LogisticRegression()
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy',n_iter=10,n_jobs=-1,random_state=42)
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```
