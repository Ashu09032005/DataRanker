```python
X = X.loc[y.index]
y.shape

```

```python
X['TestPriority'] = X['TestPriority'].str.strip().str.title()

# Then apply the mapping
priority_map = {
    'Low': 1,
    'Minor': 1,
    'Medium': 2,
    'High': 3,
    'Major': 3,
    'Very High': 3,
    'Critical': 3
}

X['TestPriority'] = X['TestPriority'].map(priority_map)

# Print to confirm
print(X)
```

```python
keep=['PASS','FAIL']
mask=y.isin(keep)
X_encoded=X.loc[mask]
table_list = table_list.loc[mask]
y=y.loc[mask]
```

```python
y_encoded=y.map({'PASS':0,'FAIL':1}).astype(int)
print(y_encoded.value_counts())
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=True, handle_unknown='ignore')
```

```python
X_encoded_sparse = encoder.fit_transform(X_encoded[[
    'TestId','CycleName','TestPriority', 'Defect1', 'Defect2', 'Defect3', 'Defect4',
    'DefectStatus1', 'DefectStatus2', 'DefectStatus3', 'DefectStatus4',
]])
```

```python
print("Length of X_encoded_sparse:", X_encoded_sparse.shape[0])
print("Length of table_list:", len(table_list))
```

```python


table_list.head()
table_list.tail()

```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded_sparse, y_encoded,test_size=0.20,random_state=42)
```

```python
##Model training

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
models={
    "Logistic Regression": LogisticRegression(multi_class="ovr"),
    "Gradient Boost":GradientBoostingClassifier()


}
average_type = 'macro'
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train, y_train)
    y_train_pred=model.predict(X_train)
    y_test_pred=model.predict(X_test)
    ## training test
    model_train_accuracy=accuracy_score(y_train, y_train_pred)
    model_train_confusion=confusion_matrix(y_train, y_train_pred)
   
    ##test performance
    model_test_accuracy=accuracy_score(y_test, y_test_pred)
    model_test_confusion=confusion_matrix(y_test, y_test_pred)
  
    print(list(models.keys())[i])
    print ("Model performance for training set data")
    print ("Accuracy: ", model_train_accuracy)
    print ("Confusion Matrix:\n", model_train_confusion)

    print("---------------------------------------")
    print ("Model performance for test set data")
    print ("Accuracy: ", model_test_accuracy)
    print ("Confusion Matrix:\n", model_test_confusion)
  
    print("---------------------------------------")
    probs=model.predict_proba(X_encoded_sparse)
    print(probs)
    print("="*32)
    probabilities = probs[:,1]

for model_name, model in models.items():
    model.fit(X_train, y_train)

    probs = model.predict_proba(X_encoded_sparse)
    print(probs)
    probabilities = probs[:, 1]

    def assign_rank(p):
        if p >= 0.8:
            return 1
        elif p >= 0.6:
            return 2
        elif p >= 0.4:
            return 3
        elif p >= 0.2:
            return 4
        else:
            return 5

    ranks = [assign_rank(p) for p in probabilities]
    
    # Create a DataFrame with the ranks
    results = pd.DataFrame({
        "Predicted_Prob": probabilities,
        "Assigned_Rank": ranks
    })
    table_list[f"Ranks by {model_name}"] = np.nan
    table_list.loc[mask, f"Ranks by {model_name}"] = ranks
```
